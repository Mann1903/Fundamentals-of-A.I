#Logistic Regression

```python
import numpy as np
data = np.array([[3.87,2.65,0],
                 [2.64,3.63,0],
                 [4.93,5.50,0],
                 [2.83,2.95,0],
                 [4.07,4.00,0],
                 [8.27,3.57,1],
                 [6.36,3.09,1],
                 [7.99,2.97,1],
                 [9.76,1.42,1],
                 [8.76,4.60,1]])

print(data)
```

```python
data[:,0:1]
```

```python
from math import exp

#Function for logistic equation output
def predict(coef,row):
  yh = coef[0] + coef[1] * row[0] + coef[2] * row[1]
  return(1 / (1+exp(-yh)))
```

```python
coef = [0.55,0.45,1.10]

for row in data:
  print('Actual target',row[2] , 'Predicted',predict(coef,row) , 'Final Prediction' , round(predict(coef,row)))
```

```python
#w0 = w0 + lr * 2 * (y-z) * z(1-z)
#w1 = w1 + lr * 2 * (y-z) * x1 * z(1-z)
#w2 = w2 + lr * 2 * (y-z) * x2 * z(1-z)

lr = 0.1

for iter in range(1000):
  sum_error = 0
  for row in data:
    z = predict(coef,row)

    coef[0] = coef[0] + lr * 2 * (row[2] - z) * z*(1-z)
    coef[1] = coef[1] + lr * 2 * (row[2] - z) * row[0] * z*(1-z)
    coef[2] = coef[2] + lr * 2 * (row[2] - z) * row[1] * z*(1-z)

    error = (row[2] - z)**2
    sum_error += error

  print(iter,sum_error)
```

```python
print(coef)
```

```python
for row in data:
  print('Actual target',row[2] , 'Predicted',predict(coef,row) , 'Final Prediction' , round(predict(coef,row)))
```

#Logistic Regression Using Library

```python
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
```

```python
df.keys()
```

```python
df['feature_names']
```

```python
import pandas as pd
X = pd.DataFrame(df['data'],columns = df['feature_names'])    #features
X
```

```python
df['target_names']
```

```python
Y = df['target']     #Labels
```

```python
#Train_test_split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y)

print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#Import Logistic Regression Algorithm
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
#The 'lbfgs' solver is a popular choice for small to medium-sized datasets in
#logistic regression. It stands for Limited-memory BroydenFletcherGoldfarbShanno.
#lmodel = LogisticRegression(solver='saga', max_iter=1000) (saga for faster convergence)

lmodel = LogisticRegression(max_iter=3000)

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on training data and transform it
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
#Train your model and evaluate it
lmodel.fit(xtrain_scaled,ytrain)
```

```python
ytrainP = lmodel.predict(xtrain_scaled)

(ytrain == ytrainP).sum()/len(xtrain_scaled)
```

```python
ytestP = lmodel.predict(xtest_scaled)

(ytest == ytestP).sum()/len(xtest_scaled)
```
