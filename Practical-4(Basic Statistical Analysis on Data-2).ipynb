**Inferential Statistics**

```python
# from IPython.display import Image
# Image('inferential.png')
```

image.png

```python
pop1=np.random.randint(10,20,1000)
np.random.seed(10)
estimates =[]
for x in range(200): # 200 samples with each 100 sample values
 sample=np.random.choice(a=pop1, size=100)
 estimates.append(sample.mean())

np.mean(pop1)
```

```python
pd.DataFrame(estimates).plot(kind='density')
```
