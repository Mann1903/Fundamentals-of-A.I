**Concatening the list**

```python
batsman=['Dhoni','Rohit','Virat','Sachin']
bowlers=['Bumrah','Shami','Chahal','Ashwin']
players=batsman + bowlers
print("Combined list of batsman and bowlres is ",players)

print("Player at index 4 is ",players[4])

players.reverse()
print("Players list in reverse order is ",players)

players[2]='pandya'
print("Updated list is ",players)
```

**Set**

```python
set1={7,2,3,4,5,6}
print("Set is ",set1)

set2={12,54,6,9,10}
print("Union of set 1 and 2 is",set1.union(set2))

print("Intersection of set is",set1.intersection(set2))

print("Difference of the set is",set1.difference(set2))
```

**Functional Programming**  

*   functions being paased as parameters to another

1.  **Map**




```python
def square(x):
  return x*x

l=[1,2,3,4]
squaredList=map(square,l)
squared=list(squaredList)
print(squared)
```

```python
intlist2 = [2,3,4,5,6]
squarelist3 = map(lambda x: x*x*x, intlist2) # Anonymous function
list(squarelist3)
```

```python
y= lambda x: x * 5
print(y(6))
print(y(8))
```

**Filter**

```python
list2=[5,6,8,15,16,20,25,26,44,45,68,67]
even = filter(lambda x: x%2 == 0, list2)
list(even)
```
