```python
pop=np.random.randint(5,25,100)
print(pop)
```

```python
#mean

np.mean(pop)
```

```python
#median

np.median(pop)
```

```python
from statistics import mode
mode(pop)
```

```python
sample=np.random.choice(pop,20)
sample

#random.choice(a, size=None, replace=True, p=None)
```

```python
sample1=np.random.choice(pop,20)
sample2=np.random.choice(pop,20)
sample3=np.random.choice(pop,20)
```

```python
all_samples=[sample1,sample2,sample3]
sample_mean=[]
for i in all_samples:
    sample_mean.append(np.mean(i))
print("Individual sample means are : ",sample_mean)
```

```python
np.mean(sample_mean)
```

**Quartiles**

```python
# from IPython.display import Image
# Image('quartile.png')
```

image.png

```python
Q1=np.percentile(n,25)
Q2=np.percentile(n,50)
Q3=np.percentile(n,75)

print(Q1,Q2,Q3)
```

**Inter Quartile Range - IQR**

```python
# from IPython.display import Image
# Image('iqr.png')
```

image.png

```python
IQR=Q3-Q1
print(IQR)
```

**Outlier**

```python
Outlier=Q1- 1.5 * IQR, Q3+1.5*IQR
print(Outlier)
```
